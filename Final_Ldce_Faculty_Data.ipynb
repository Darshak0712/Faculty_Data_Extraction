```python
def get_all_faculty_details(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content , 'html.parser')
    
    Name = soup.find('h1',{'itemprop':'name'})
    Designation = soup.find('td',{'itemprop':'jobTitle'})
    Mail = soup.find('td',{'itemprop':'email'})
    Department = soup.find('td',{'itemprop':'memberof'})
    
    print("_______________________________________________________________________________________________________\n")
    print(f'Name : {Name.text}')
    print(f'Designation : {Designation.text}')
    print(f'Mail-ID : {Mail.text}')
    print(f'Department : {Department.text}')
    print(f"Profile-Link : {url}")
    
    data = soup.find_all(class_='panel panel-profile')
    
    for i in range(len(data)):
        
        
        title = data[i].find(class_="panel-title heading-sm pull-left").text
        
        
        
        if title == 'Social Contacts'or title == 'Paper Publication':
            continue
            
        print(f"\n-------{title}--------")
        
        post_year_lst = data[i].find_all('li')
        
        for more_details in post_year_lst:
            
            post_year = more_details.find_all('span')
            
            place = more_details.find('h2')
            
            if None in (post_year,place):
                continue
            
            print(f"Details : {post_year[0].text} ( {post_year[1].text} ) | {place.text}\n")
#             print(f"Place : {place.text}\n")
    
        
        
```
